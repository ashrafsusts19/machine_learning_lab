```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import itertools



dset = pd.read_csv("/content/sample_data/bank-full.csv", sep=";")


X = dset.drop(columns = "y", axis = 1)
Y = dset['y']
# print(X)
# print(Y)
dset.head()
```

```python

label_encode = LabelEncoder()
enc_job = label_encode.fit_transform(X.job)
X['job'] = enc_job
enc_marital = label_encode.fit_transform(X.marital)
X['marital'] = enc_marital
enc_education = label_encode.fit_transform(X.education)
X['education'] = enc_education
enc_default = label_encode.fit_transform(X.default)
X['default'] = enc_default
enc_housing = label_encode.fit_transform(X.housing)
X['housing'] = enc_housing
enc_loan = label_encode.fit_transform(X.loan)
X['loan'] = enc_loan
enc_contact = label_encode.fit_transform(X.contact)
X['contact'] = enc_contact
enc_month = label_encode.fit_transform(X.month)
X['month'] = enc_month
enc_poutcome = label_encode.fit_transform(X.poutcome)
X['poutcome'] = enc_poutcome
enc_y = label_encode.fit_transform(Y)
Y = enc_y
X.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)

scaler = StandardScaler()
scaler.fit(X_train)

X_train_standardized = scaler.transform(X_train)
X_test_standardized = scaler.transform(X_test)

w = np.ones(16)
print(X_train.dot(w))

# rows are individiual data, columns are field value
# X_train_standardized.shape
```

```python
class Logistic_Regression:
  alpha = 0.01
  iteration_count = 1000
  def __init__(self, alpha, iteration_count):
    self.alpha = alpha
    self.iteration_count = iteration_count

  def fit(self, X, Y):
    self.rows, self.features = X.shape

    self.w = np.zeros(self.features)
    self.b = 0
    self.X = X
    self.Y = Y
    for _ in range(self.iteration_count):
      self.updateWeights()

  def updateWeights(self):
    z = self.X.dot(self.w) + self.b

    Y_hat = 1 / (1 + np.exp(-z))

    dw = (1 / self.rows) * np.dot(self.X.T, (Y_hat - self.Y))
    db = (1 / self.rows) * np.sum(Y_hat - self.Y)

    self.w = self.w - self.alpha * dw
    self.b = self.b - self.alpha * db

  def predict(self, testX):
    z = testX.dot(self.w) + self.b

    Y_pred = 1 / (1 + np.exp(-z))
    Y_pred = np.where(Y_pred > 0.5, 1, 0)
    return Y_pred


model = Logistic_Regression(0.01, 1000)
model.fit(X_train, Y_train)

X_test2 = X_test.to_numpy()


cnt_cor, cnt_incor = 0, 0
for i in range(len(Y_test)):
  model_result = model.predict(X_test2[i])
  actual_result = Y_test[i];
  if model_result == actual_result:
    cnt_cor += 1;
  else:
    cnt_incor += 1;

print(cnt_cor, cnt_incor)
print(cnt_cor / (cnt_cor + cnt_incor))
```
